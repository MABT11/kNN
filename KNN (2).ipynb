First we will import the modeules

```python
import pandas as pd
import numpy as np
```

Read the data

```python
df = pd.read_csv("https://filedn.com/lYFhTM6v23XRGBo3YevN005/breast_cancer.txt", delim_whitespace=True, header=None)
```

Display the shape of the data

```python
df.shape
```

To Display the first 25 rows

```python
df.head(25)
```

To Extract the first column of the dataframe into a new data frame called “target”.

```python
target = df.iloc[:,0]
target
```

Count the frequency of the two classes (-1 and 1) in “target”

```python
index = pd.Index(target)
print(index.value_counts())
```

Extracting all the remaining 30 columns of “df” into a new data frame called “features”.

```python
# [row_start:row_end , column_start, column_end]
features = df.iloc[:,1:-1]
features
```

Display the statistics of “features” using the describe() function.

```python
features.describe()
```

The columns of “features” are of different scales. So we convert them into a uniform scale (0 .. 1) and create a new data frame “features_scaled”.

```python
# we first import the MinMaxScaler to scale the data
from sklearn.preprocessing import MinMaxScaler
# apply the scaler and transform the numpy array to dataframe to make it more readable
features_scaled = pd.DataFrame(MinMaxScaler().fit_transform(features),columns=features.columns)
```

The first 5 rows of “features_scaled”.

```python
features_scaled.head(5)
```

The statistics of “features_scaled” are displayed

```python
features_scaled.describe()
```

Using the train_test_split() function, we split “features_scaled” and “target” into the four new data frames: “X_train”, “X_test”, “y_train”, and “y_test”. (60% for training and 40% for testing.)

```python
# import the train-test-split
from sklearn.model_selection import train_test_split
# call the train-test-split witht the features_scaled df
x_train,x_test,y_train,y_test = train_test_split(features_scaled,target,test_size=0.4,random_state=123)
```

Getting the shapes and data types of “X_train”, “X_test”, “y_train”, and “y_test”.  

```python
print(f"X_train = {x_train.shape}")
print(f"X_test = {x_test.shape}")
print(f"Y_train = {y_train.shape}")
print(f"Y_test = {y_test.shape}")
```

Using KNN, We declare a KNN model with 1 nearest neighbor.

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=1)
```

Fitting “X_train” and “y_train” using the KNN model.

```python
knn.fit(x_train,y_train)
```

Using the fitted kNN model we predict the class labels of “X_test”.

```python
y_predict = knn.predict(x_test)
```

• Comparing “y_test” and “y_predict” to spot the prediction errors.

        ◦ To compare “y_predict” (a list) with “y_test” (a Pandas series), we need to create a list “y_test_list” from the latter. (using the list() function)
        ◦ If there is a prediction error for a test sample, we will display the test sample ID, actual label, and predicted label.

```python
y_test_list = list(y_test)
for i in range(len(y_test_list)):
  if y_test_list[i] != y_predict[i]:
      print(f"The row number is {i}, the actual label is {y_test_list[i]} and the predicted label is {y_predict[i]}")      
```

Display the test accuracy

```python
from sklearn.metrics import accuracy_score
acc = accuracy_score(y_test,y_predict)
print(acc)
```

•	We Repeat the above kNN process for 3, 5, 7, 9, and 11 nearest neighbors.

```python
e = []
for i in range(3,13,2):
  knn = KNeighborsClassifier(n_neighbors=i)
  knn.fit(x_train,y_train)
  predictions = knn.predict(x_test)
  e.append(np.mean(y_test!=predictions))
min = e[0]
index = 0
for i in range(len(e)):
  print(f"K = {2*i+3} the error is {e[i]}")
  if min > e[i]:
    min = e[i]
    index = 2*i+3

print(f"The lowest error is {min} when k = {index}")
```

So when we choose k=7 we will get the least amount of error. As demonstrated below by the graph also.

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8,4))
plt.plot(range(3,13,2),e,color='blue', marker='o',linestyle='--')
plt.xlabel('Neighbors')
plt.ylabel('error')
```
